## Observations and Insights 

1) Based on the data presented, Capomulin and Ramicane should be used most going forward. While having the most data collected, the mice they treated had comparitavely much smaller final tumor sizes than thie competitors. While it is possible the other drugs were treating mice with larger initial tumor sizes, it seems that the starting mice were comparable, and thus, these treatments were more effective.

2) Supporting the above, per our observation of a specific mouse, Capomulin regimen directly relates to the decrease of tumor size over time. 

3) There is an almost perfect correlation between mouse weight and tumor volume amongst those treated with Capomulin. From glancing at the data, it does not appear the study showed significant (if any) weight gain or loss over treatment time. This adds another layer to the above insights, as it would appear that the treatment works in shrinking the tumor without causing significant weight gain or loss.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
# Add NumPy
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
all_data = pd.merge(mouse_metadata,study_results, on="Mouse ID")
# Display the data table for preview
all_data
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
lasttime = clean_data.groupby("Mouse ID")["Timepoint"].max()

# Merge this Grouped DataFrame with the original DataFrame to assign the tumor volume at the last timepoint to each line of each mouse
new_clean = pd.merge(clean_data, lasttime,  on = "Mouse ID", how = "left")
# Rename the columns as Timepoint and Max Timepoint for clarity
new_clean = new_clean.rename(columns={"Timepoint_x":"Timepoint", "Timepoint_y":"Max Timepoint"})
# Create new DataFrame that only keeps the line for the mouse at the last timepoint
max_only = new_clean.loc[(new_clean["Timepoint"]==new_clean["Max Timepoint"])]
# Reset the index to begin at 0 again
max_only = max_only.reset_index(drop=True)
# Drop the Max Timepoint column, as the Timepoint column now represents the last timepoint
max_only = max_only.drop(columns={"Max Timepoint"})
# Display the new DataFrame
max_only
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
capvol = []
ramvol = []
infvol = []
cefvol = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

    
    # Locate the rows which contain mice on each drug and get the tumor volumes

# Create "for loop" to place the volume associated with each treatment in the lists defined above
for x in max_only.index:
    if max_only["Drug Regimen"][x] == treatments[0]: 
        capvol.append(max_only["Tumor Volume (mm3)"][x])
    if max_only["Drug Regimen"][x] == treatments[1]:
        ramvol.append(max_only["Tumor Volume (mm3)"][x])
    if max_only["Drug Regimen"][x] == treatments[2]:
        infvol.append(max_only["Tumor Volume (mm3)"][x])
    if max_only["Drug Regimen"][x] == treatments[3]:
        cefvol.append(max_only["Tumor Volume (mm3)"][x])
    
    # add subset 
allvol = [capvol, ramvol, infvol, cefvol]

# add lists to DataFrame for easy viewing
tumvol = pd.DataFrame ({"Capomulin":capvol,
                       "Ramicane":ramvol,
                       "Infubinol":infvol,
                       "Ceftamin":cefvol})
     # Determine outliers using upper and lower bounds
# Define each quartile. This DataFrame will show each quartile as a row with the Drug Regimens as columns
quartiles = tumvol.quantile([.25,.5,.75])
# Capture the lower and upper quartiles of the DataFrame
lowerq = quartiles.iloc[0,:]
upperq = quartiles.iloc[2,:]
# Caluclate the IQR and bounds
iqr = upperq - lowerq
lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)
# Code idea to count outliers came from Stack Overflow (https://stackoverflow.com/questions/39068214/how-to-count-outliers-for-all-columns-in-python)
outliers = ((tumvol < lower_bound) | (tumvol > upper_bound)).sum()
# Use this to create a line break in the print statement below
nl = '\n'
# Print the outliers list
print(f"Here are the number of outliers by Drug Regimen:{nl}{nl}{outliers}")    
# Knowing that the outlier is in Infubinol, search for the value of the outlier
outlier_df = tumvol.loc[(tumvol["Infubinol"] > upper_bound[2]) | (tumvol["Infubinol"] < lower_bound[2]), ["Infubinol"]]
# Print the outlier value
print(f'The value of the outlier is: {nl}{outlier_df.to_string(index=False, header=False)}')
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
# Create subplots to allow multiple boxplots in one figure
fig1, ax1 = plt.subplots()
# Create labels for axes/title
ax1.set_title('Tumor Volumes by Drug Regimen')
ax1.set_ylabel('Tumor Volume (mm3)')
# Create boxplot, have fun with formatting!
ax1.boxplot(allvol, labels = treatments, flierprops=dict(markerfacecolor='r', marker='d'), patch_artist = True, boxprops=dict(facecolor= "yellow", color="black"), capprops = dict(color = "black"), medianprops = dict(color = "black"))
# Color background of plot for fun!
ax1.set_facecolor("lightgreen")
# Display plot
plt.show()

```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
# Group all data by Mouse ID and then by Timepoint, and sum the data (will present the actual volume at given timepoint)
line_groups = clean_data.groupby(['Mouse ID', 'Timepoint']).sum()
# Create variable for the mouse to use
cap_mouse = "s185"
# Create a filtered DataFrame with only mouse s185's info
just_one_mouse = line_groups.loc[cap_mouse]
# Create line plot for the selected mouse
mouse_line = just_one_mouse.plot(kind="line", y='Tumor Volume (mm3)', title="Mouse s185 Response to Capomulin", grid = True, xlim = (0,50), ylim= (20,50), linewidth = 5)
# Add label to y-axis
mouse_line.set_ylabel("Tumor Volume (mm3)")
# Display plot
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
# Group all data by Drug Regimen and then Weight, then take the average of other columns (will return avg volume by weight)
scat_groups =  clean_data.groupby(['Drug Regimen','Weight (g)']).mean()
# Create a filtered DataFrame with only mice treated by Capomulin
scat_reg = scat_groups.loc["Capomulin"]
# Reset the index so that Weight can be used as a column
scat_reg = scat_reg.reset_index()
# Create scatter plot
scat_plot = scat_reg.plot(kind="scatter", x="Weight (g)", y="Tumor Volume (mm3)", title="Avg. Tumor Volume by Weight of Mice Treated With Capomulin", s = 60, xlim= (14,26), ylim= (34,46))
# Add label to y-axis
scat_plot.set_ylabel("Average Tumor Volume (mm3)")
# Display plot
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
# Using the Grouped DF above, create variables to be used in correlation and as axes in plotting
cap_weight = scat_reg["Weight (g)"]
cap_vol = scat_reg["Tumor Volume (mm3)"]
# Calculate correlation coefficient and print statement
correlation = st.pearsonr(cap_weight,cap_vol)
print(f"The correlation between both factors is {round(correlation[0],2)}")

# Using the linregress module from Scipy Stats, assign values to variables related to linear regression
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(cap_weight,cap_vol)
# Create variable for the formula serving as y=mx+b for the linear regression
regress_values = cap_weight * slope + intercept
# Create variable for the text to explain the y=mx+b formula
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
# Create scatter plot
plt.scatter(cap_weight,cap_vol, s = 60)
# Plot the line "regress_values" to visually represent the linear regression
plt.plot(cap_weight,regress_values,"r-")
# Plot the formula of the line in text
plt.annotate(line_eq,(18,37),fontsize=15,color="red")
# Plot the labels for axes/title
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title("Avg. Tumor Volume by Weight of Mice Treated With Capomulin")
# Set the limits of the plot
plt.xlim(14,26)
plt.ylim(34,46)
# Display the plot
plt.show()
```
